```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.sample(5)
```

# Categorical Data 

## a. countplot

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.catplot(x='Pclass',kind='count',data=df)
```

```python
sns.catplot(x='Sex',kind='count',data=df)
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
MaleFemale=df['Sex'].value_counts()
```

```python
plt.pie(MaleFemale,labels=['male','female'],autopct='%0.2f')
plt.show()
```
